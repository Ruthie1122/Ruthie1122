```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def balanc(dataframe):
    # Ensure the 'Class' column exists in the dataframe
    if 'Class' not in dataframe.columns:
        print("Column 'Class' not found in the dataframe.")
        return

    # Get the value counts of each class
    value_count = dataframe['Class'].value_counts()

    # Check if both classes are present
    if len(value_count) != 2:
        print("The dataset does not contain both classes.")
        return

    # Calculate the total number of samples
    total = value_count.sum()

    # Calculate the percentage of each class
    fraud_percentage = (value_count[1] / total) * 100
    non_fraud_percentage = (value_count[0] / total) * 100

    # Print the results
    print(f"No Frauds: {non_fraud_percentage:.2f}% of the dataset")
    print(f"Frauds: {fraud_percentage:.2f}% of the dataset")

def plot_balanc(data=None, column='Class'):
    # Ensure the data is provided and column exists
    if data is None or column not in data.columns:
        print(f"Column '{column}' not found in the provided data.")
        return

    # Create a count plot for the specified column
    sns.countplot(data=data, x=column)
    plt.title('Class Distributions \n (0: No Fraud || 1: Fraud)', fontsize=14)
    plt.xlabel('Class')
    plt.ylabel('Count')
    plt.show()

balanc(dataframe)
plot_balanc(data=dataframe, column='Class')
```

```python
import numpy as np

def prep_data(dataframe):
    # Drop the 'Class' column and store the rest as features
    X = dataframe.drop(['Class'], axis=1).values.astype(float)
    # Extract the 'Class' column as the target
    y = dataframe[['Class']].values.astype(float).flatten()
    return X, y

# Create X and y from the prep_data function
X, y = prep_data(dataframe)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler, SMOTE, BorderlineSMOTE

# Assuming X and y are already defined and preprocessed

# Create the training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Define the resampling methods
undersam = RandomUnderSampler(random_state=0)
oversam = RandomOverSampler(random_state=0)
smote = SMOTE(random_state=0)
borderlinesmote = BorderlineSMOTE(random_state=0)

# Resample the training data
X_undersam, y_undersam = undersam.fit_resample(X_train, y_train)
X_oversam, y_oversam = oversam.fit_resample(X_train, y_train)
X_smote, y_smote = smote.fit_resample(X_train, y_train)
X_borderlinesmote, y_borderlinesmote = borderlinesmote.fit_resample(X_train, y_train)

# Display the resampled dataset shapes
print("Original dataset shape:", np.bincount(y_train.astype(int)))
print("Undersampled dataset shape:", np.bincount(y_undersam.astype(int)))
print("Oversampled dataset shape:", np.bincount(y_oversam.astype(int)))
print("SMOTE dataset shape:", np.bincount(y_smote.astype(int)))
print("BorderlineSMOTE dataset shape:", np.bincount(y_borderlinesmote.astype(int)))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Create the training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=40)

# Fit a logistic regression model to our data
model = LogisticRegression()
model.fit(X_train, y_train)

# Obtain model predictions
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import roc_curve,roc_auc_score, precision_recall_curve, average_precision_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

# Create true and false positive rates
false_positive_rate, true_positive_rate, threshold = roc_curve(y_test, y_predicted)

# Calculate Area Under the Receiver Operating Characteristic Curve
probs = model.predict_proba(X_test)
roc_auc = roc_auc_score(y_test, probs[:, 1])
print('ROC AUC Score:',roc_auc)

# Obtain precision and recall
precision, recall, thresholds = precision_recall_curve(y_test, y_predicted)

# Calculate average precision
average_precision = average_precision_score(y_test, y_predicted)

# Print the classifcation report and confusion matrix
print('Classification report:\n', classification_report(y_test, y_predicted))
print('Confusion matrix:\n',confusion_matrix(y_true = y_test, y_pred = y_predicted))
```

```python
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import precision_recall_curve, auc, classification_report

# Scale 'Time' and 'Amount'
scaler = StandardScaler()
dataframe[['Time', 'Amount']] = scaler.fit_transform(dataframe[['Time', 'Amount']])

# Prepare data
X = dataframe.drop('Class', axis=1)
y = dataframe['Class']

# Check class distribution
print("Class distribution before SMOTE:")
print(y.value_counts())

# Determine the number of samples in the minority class
minority_class_count = y.value_counts().min()

# Set k_neighbors to be less than or equal to the number of samples in the minority class
k_neighbors = min(5, minority_class_count - 1)  # Adjust the default 5 to be less than or equal to minority_class_count - 1

# Apply SMOTE
sm = SMOTE(random_state=42, k_neighbors=k_neighbors)
X_res, y_res = sm.fit_resample(X, y)

# Check class distribution after SMOTE
print("Class distribution after SMOTE:")
print(y_res.value_counts())

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
```

The project focused on detecting fraud in a financial dataset using machine learning techniques, with a specific emphasis on handling class imbalance. Here are the key components and findings of the project:

1. Data Preprocessing and Exploration
Dataset Characteristics: The dataset contained multiple features with a significant class imbalance, where fraudulent transactions were much fewer than non-fraudulent ones.
Exploratory Data Analysis (EDA): This step involved understanding the distribution of data, visualizing the class imbalance, and examining correlations among features.
2. Handling Class Imbalance
Various resampling techniques were applied to address the class imbalance:
Random Under-Sampling: Reduced the number of non-fraudulent transactions to balance the classes, potentially leading to information loss.
Random Over-Sampling: Increased the number of fraudulent transactions by duplicating existing samples, which may lead to overfitting.
SMOTE (Synthetic Minority Over-sampling Technique): Created synthetic samples for the minority class, providing a more balanced dataset without exact duplicates.
BorderlineSMOTE: Focused on generating synthetic samples near the decision boundary, enhancing the model's ability to distinguish between classes.
3. Model Training and Evaluation
Model Selection: Various models, including Logistic Regression, Random Forest, and XGBoost, were evaluated.
Cross-Validation: A 5-fold cross-validation approach was used to assess the model's performance across different subsets of the data, providing insights into generalization capabilities.
4. Performance Metrics
ROC AUC Score: The models achieved varying levels of performance, with the best model achieving a high ROC AUC score, indicating good discrimination between fraudulent and non-fraudulent transactions.
Classification Metrics: Precision, recall, F1-score, and confusion matrix provided a detailed view of the model's performance, highlighting the challenges in correctly identifying fraudulent transactions.
5. Challenges and Considerations
Class Imbalance: The severe class imbalance was a central challenge, impacting model performance and necessitating careful handling through resampling techniques.
Variability in Cross-Validation Scores: The significant drop in performance on one fold suggested potential issues such as data noise, class distribution differences, or overfitting.
Model Stability: Ensuring consistent performance across all data subsets remains a critical concern, highlighting the need for robust validation strategies.
